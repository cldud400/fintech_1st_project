```python
import pandas as pd
import numpy as np
```

```python
import warnings

warnings.filterwarnings(action='ignore')
```

```python
# LIT ETF

df_lit = pd.read_csv('./Data/리튬 etf LIT.csv')
df_lit
```

```python
df_lit.drop(['Open', 'High', 'Low', 'Close'], axis = 1, inplace=True)
df_lit
```

```python
df_lit.columns = 'date', 'LIT조정종가', 'LIT거래량'
df_lit
```

```python
df_lit['date'] = pd.to_datetime(df_lit['date'])
```

```python
df_lit.info()
```

```python
# SGL ETF

df_sgl = pd.read_csv('./Data/리튬전지 SGL etf.csv')
df_sgl
```

```python
df_sgl.drop(['오픈', '고가', '저가', '거래량'], axis = 1, inplace = True)
df_sgl
```

```python
df_sgl.columns = 'date', 'SGL종가', 'SGL변동률'
df_sgl
```

```python
df_sgl['date'] = df_sgl['date'][0].replace('년 ', '-')
```

```python
df_sgl['date'] = df_sgl['date'][0].replace('월 ', '-')
```

```python
df_sgl['date'] = df_sgl['date'][0].replace('일', '')
```

```python
df_sgl['date']
```

```python
df_sgl['date'] = pd.to_datetime(df_sgl['date'])
```

```python
# merge

pd.merge(df_lit, df_sgl, on = 'date', how = 'outer')
```
