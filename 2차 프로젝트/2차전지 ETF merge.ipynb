```python
import pandas as pd
import numpy as np
```

```python
def m_to_k(x):
    if x[-1] == 'M':
        return float(x[:-1]) * 1000
    elif x[-1] == 'K':
        return float(x[:-1])
    elif x[-1] == 'B':
        return float(x[:-1]) * 1000000
    else:
        return np.nan
```

```python
def change_date(x):
    ret = []
    for da in x.split():
        ret.append(da[:-1])
    return '.'.join(ret)
```

```python
# 타이거 2차전지 테마 ETF

df_tiger = pd.read_csv('./Data/305540 타이거 2차전지테마(ETF).csv')
df_tiger
```

```python
df_tiger.drop(['오픈', '고가', '저가'], axis = 1, inplace = True)
```

```python
df_tiger.columns = 'date', 'tiger종가', 'tiger거래량', 'tiger변동률'
```

```python
df_tiger['date'] = pd.to_datetime(df_tiger['date'].apply(change_date))
```

```python
df_tiger['tiger거래량'] = df_tiger['tiger거래량'].apply(m_to_k)
```

```python
df_tiger['tiger변동률'] = df_tiger['tiger변동률'].apply(lambda x : x[:-1]).astype('float')
```

```python
df_tiger['tiger종가'] = df_tiger['tiger종가'].apply(lambda x : x.replace(',', '')).astype('float')
```

```python
# Kodex 2차전지산업 ETF

df_kodex = pd.read_csv('./Data/한국 2차전지 etf KODEX 2차전지산업.csv')
df_kodex
```

```python
df_kodex.drop(['오픈', '고가', '저가'], axis = 1, inplace = True)
```

```python
df_kodex.columns = 'date', 'kodex종가', 'kodex거래량', 'kodex변동률'
```

```python
df_kodex['date'] = pd.to_datetime(df_kodex['date'].apply(change_date))
```

```python
df_kodex['kodex변동률'] = df_kodex['kodex변동률'].apply(lambda x : x.replace('%', '')).astype('float')
```

```python
df_kodex['kodex종가'] = df_kodex['kodex종가'].apply(lambda x : x.replace(',', '')).astype('float')
```

```python
df_kodex['kodex거래량'] = df_kodex['kodex거래량'].apply(m_to_k).astype('float')
```
