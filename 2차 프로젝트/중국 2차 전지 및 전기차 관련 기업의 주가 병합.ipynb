```python
import pandas as pd
import numpy as np
```

```python
import warnings

warnings.filterwarnings(action='ignore')
```

```python
def m_to_k(x):
    if x[-1] == 'M':
        return float(x[:-1]) * 1000
    elif x[-1] == 'K':
        return float(x[:-1])
    elif x[-1] == 'B':
        return float(x[:-1]) * 1000000
    else:
        return np.nan
```

```python
def change_date(x):
    ret = []
    for da in x.split():
        ret.append(da[:-1])
    return '.'.join(ret)
```

```python
# EVE Energy 주가

df_eve = pd.read_csv('./Data/300014 EVE Energy 주가.csv')
df_eve
```

```python
df_eve.drop(['오픈', '고가', '저가'], axis = 1, inplace = True)
df_eve.head()
```

```python
df_eve.columns = 'date', 'EVE종가', 'EVE거래량', 'EVE변동률'
```

```python
df_eve['date'] = pd.to_datetime(df_eve['date'].apply(change_date))
```

```python
df_eve['EVE거래량'] = df_eve['EVE거래량'].apply(m_to_k).astype('float')
```

```python
df_eve['EVE변동률'] = df_eve['EVE변동률'].apply(lambda x : x[:-1]).astype('float')
```
