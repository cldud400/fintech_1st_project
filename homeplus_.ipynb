```python
# crawling 검색해서 dataframe만들기
df_res = pd.DataFrame(columns = ['상품명', '가격'], index = [0])
driver = webdriver.Chrome()   

import time
cnt = 0
for pro in df.상품명.unique():
    
    pro = pro.replace("개입"," ").replace("개"," ").replace("매"," ").replace("("," ").replace(")"," ").replace("*"," ").replace("캔"," ").replace("팩"," ").replace("세트"," ").replace("21cm x 16cm x 6cm"," ")        
    url = 'https://front.homeplus.co.kr/search?entry=direct&keyword={}'.format(pro)
    driver.get(url)
   
    time.sleep(0.15)
    path = 'div.itemDisplayList'
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    prod = soup.select(path)

    
    if prod:
        for i in prod:
            # df_res = df_res[df_res['상품명'].str.contains('x' , na=False, case=False)]
            name = i.select('div.detailInfo > a > p.css-1tf7nrf-defaultStyle-Typography-ellips')[0].text
            price = i.select('strong')[0].text
            df_res = df_res.append({'상품명' :name ,'가격' : price},ignore_index=True)
            break
        else:
            df_res = df_res.append({'상품명' :pro ,'가격' : np.nan},ignore_index=True)
```

```python
# homeplus dataframe 전처리
df_homeplus = pd.read_csv('homeplus.csv')
df_homeplus.drop('Unnamed: 0', axis=1, inplace=True)
df_homeplus['가격'] = df_homeplus['가격'].apply(lambda x : x.replace(',',''))
df_homeplus['가격'] = df_homeplus['가격'].astype('int')
```
